```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Define the features (X) and target variable (y)
X = encoded_df.drop(columns=['price'])
y = encoded_df['price']

# Initialize KFold with k=5 (you can adjust k as needed)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize a list to store the mean squared errors for each fold
mse_scores = []

# Iterate through each fold
for train_index, test_index in kf.split(X):
    # Split the data into train and test sets for this fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Initialize and fit the model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Calculate mean squared error and append to the list
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Calculate the mean of the MSE scores
mean_mse = np.mean(mse_scores)
print("Mean Squared Error:", mean_mse)
```

```python
from sklearn.model_selection import KFold, cross_val_score

def train_and_evaluate_model_with_cv(df, model_type='rf', target_column='price', n_splits=5, random_state=42, n_estimators=100):
    # Extract features and target variable
    X = df.drop(target_column, axis=1)
    y = df[target_column]

    # Initialize the model based on the model_type parameter
    if model_type == 'rf':
        model = RandomForestRegressor(n_estimators=n_estimators, random_state=random_state)
    elif model_type == 'svm':
        model = SVR()
    elif model_type == 'lr':
        model = LinearRegression()
    else:
        raise ValueError("Invalid model type. Please choose from 'rf', 'svm', or 'lr'.")

    # Define the cross-validation strategy
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    # Perform k-fold cross-validation
    mse_scores = cross_val_score(model, X, y, cv=kf, scoring='neg_mean_squared_error')

    # Convert negative mean squared error scores to positive values
    mse_scores = -mse_scores

    # Calculate the mean and standard deviation of MSE scores
    mean_mse = mse_scores.mean()
    std_mse = mse_scores.std()

    print("Mean Squared Error ({} with {}-fold CV): {:.3f} +/- {:.3f}".format(model_type.upper(), n_splits, mean_mse, std_mse))

    return model, mean_mse

# Example usage:i thin
# Assuming 'encoded_df' contains your dataset and 'price' is the target variable
model_types = ['rf', 'svm', 'lr']
for model_type in model_types:
    trained_model, mean_mse = train_and_evaluate_model_with_cv(encoded_df, model_type=model_type)
```
