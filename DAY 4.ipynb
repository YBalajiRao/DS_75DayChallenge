```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.01
    costs = []

    for i in range(iterations):
        y_prediction = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y - y_prediction)])
        md = -(2/n) * sum(x * (y - y_prediction))
        bd = -(2/n) * sum(y - y_prediction)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        costs.append(cost)

        if i % 100 == 0:
            print("Iteration {}: m {}, b{}, cost{}".format(i, m_curr, b_curr, cost))

    # Plotting the regression line and data points
    plt.scatter(x, y, color='red', marker='o')
    plt.plot(x, m_curr * x + b_curr, color='blue')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Linear Regression Fit')
    plt.show()



x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
gradient_descent(x, y)
```
